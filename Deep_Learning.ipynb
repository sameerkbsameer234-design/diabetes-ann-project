```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import warnings
warnings.filterwarnings('ignore')
```

```python
diabetes = load_diabetes()

x = diabetes.data
y = diabetes.target

df = pd.DataFrame(x, columns=diabetes.feature-names)
df['target'] = y

df.head()
```

```python
print("Missing Values:\n", df.isnull().sum())
```

# Feature Scaling

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

# EDA-Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(df['target'], bins=30, kde=True)
plt.title("Target Variable Distribution")
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

# Feature vs Target

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['bmi'], y=df['target'])
plt.title("BMI vs Diabetes Progression")
plt.show()
```

# Train-Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y, test_size=0.2, random_state=42
)
```

# Build ANN Model

```python
model = Sequential()

model.add(Dense(64, activation='relu', input_shape=(x_train.shape[1],)))

model.add(Dense(32, activation='relu'))

model.add(Dense(1))

model.summary()
```

# Compile Model

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

# Train Model

```python
history = model.fit(
    x_train, y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    verbose=1
)
```

# Evaluate Model

```python
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

# Improved Model

```python
model2 = Sequential()

model2.add(Dense(128, activation='relu', input_shape=(x_train.shape[1],)))
model2.add(Dense(64, activation='relu'))
model2.add(Dense(32, activation='relu'))
model2.add(Dense(1))

model2.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']
)

history2 = model2.fit(
    x_train, y_train,
    epochs=150,
    batch_size=16,
    validation_split=0.2,
    verbose=1
)
```

# Evaluate Improved Model

```python
y_pred2 = model2.predict(x_test)

mse2 = mean_squared_error(y_test, y_pred2)
r2_2 = r2_score(y_test, y_pred2)

print("Improved Model MSE:", mse2)
print("Improved Model R2:" r2_2)
```

# Compare Results

```python
print("Before Improvement:")
print("MSE:", mse)
print("R2:", r2)

print("\nAfter Improvement:")
print("MSE:", mse2)
print("R2:", r2_2)
```
